```python
import matplotlib.pyplot as plt

import numpy as np
```

### Multiply

```python
my_x = [0, 1, 1]
my_w = [-10, 10, 5]
```

```python
def mul(a, b):
    return [i * j for i, j in zip(a, b)]
```

```python
mul(my_x, my_w)
```

### Linear Neuron

```python
my_b = -5
```

```python
def linear_neuron(x, weights, bias):
    return sum(mul(x, weights)) + bias
```

```python
linear_neuron(my_x, my_w, my_b)
```

### Sigmoid

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 100)
c1s = [.5, 1, 2]
c2s = [-1, 0, 1]
for c1 in c1s:
    plt.plot(x, sigmoid(c1*x), label=c1)
plt.legend()
    
plt.figure()
for c2 in c2s:
    plt.plot(x, sigmoid(x + c2), label=c2)
plt.legend()
```

```python
def neuron(x, weights, bias):
    return sigmoid(sum(mul(x, weights)) + bias)
```

```python
neuron(my_x, my_w, my_b)
```
